```python
from pathlib import Path
from sklearn import metrics
from collections import Counter

path_pred = Path("/afs/crc.nd.edu/group/dmsquare/vol2/myu2/ComparisonSentences/experiments/pairwise_cls/sbert_combined_cross_0/predictions.txt")
references = []
predictions = []
with open (path_pred) as f:
    for line in f:
        arr = line.strip().split('\t')
        label, sp, sentence_1, sentence_2 = arr[0].split('_')
        score = float(arr[1])
        predictions.append(score)
        if label == "pos":
            references.append(1)
        elif label == "neg":
            references.append(0)

print(Counter(references))
print(max(predictions), min(predictions))
print(references[:10])
print(predictions[:10])
assert len(references) == len(predictions)

fpr, tpr, thresholds = metrics.roc_curve(references, predictions, pos_label=1)
print(fpr)
print(tpr)
print(thresholds)
auc = metrics.auc(fpr, tpr)
print(auc)

import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve
precisions, recalls, thresholds = precision_recall_curve(references, predictions)

plt.plot(thresholds, precisions[:-1])
plt.plot(thresholds, recalls[:-1])
plt.show()

plt.plot(precisions, recalls)
plt.show()

f1s = (2 * precisions * recalls) / (precisions + recalls)
print("max f1", max(f1s))
plt.plot(thresholds, f1s[:-1])
plt.show()
```
