```python
import json
path_ref = "/afs/crc.nd.edu/group/dmsquare/vol2/myu2/ComparisonSentences/data/sentence_scoring/v2/Q5/pairwise_cls/cross_1/pairwise_test.json"
path_hyp = "/afs/crc.nd.edu/group/dmsquare/vol2/myu2/ComparisonSentences/experiments/pairwise_cls/cross_enc_train100k_split1_1117/predictions.txt"

refs = []
with open(path_ref) as f:
    for line in f:
        refs.append(json.loads(line))

preds = []
with open(path_hyp) as f:
    next(f)
    for line in f:
        preds.append(line.strip().split('\t'))
    
cnt = 0
for i in range(len(refs)):
    assert refs[i]['id'] == preds[i][0]
    if refs[i]['label'] != int(preds[i][1]):
        print(refs[i])
        cnt += 1
print(cnt)
```
