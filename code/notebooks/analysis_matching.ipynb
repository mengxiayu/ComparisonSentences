```python
import json
from pathlib import Path
from collections import Counter
dir_matched = Path("/afs/crc.nd.edu/group/dmsquare/vol2/myu2/ComparisonSentences/data/statement_scoring/news_v1")

def analysis_matched(dir_matched):
    cnt_etype = 0
    for fn_etype in dir_matched.glob("*matched.json"):
        etype = fn_etype.stem.rstrip("_matched.json")
        if cnt_etype == 10:
            break
        cnt_etype += 1
        
        print('\n', etype)
        p2freq = Counter()
        p2alias = {}
        with open(fn_etype) as f:
            for line in f:
                try:
                    obj = json.loads(line)
                except:
                    # print(line)
                    continue
                e1, e2 = obj["entity_pair"]
                p = obj["property"]
                p2freq[p] += 1
                ids, aliases, sent = obj["evidence_e1"][0]
                if p not in p2alias:
                    p2alias[p] = aliases[1]
                
        for k, v in p2freq.most_common(5):
            print(k, p2alias[k], v)
            
                
```

```python
# clean files
for fn_etype in dir_matched.glob("*matched.json"):
    etype = fn_etype.stem.rstrip("_matched.json")
    all_lines = []
    with open(fn_etype) as f:
        for line in f:
            try:
                obj = json.loads(line)
                all_lines.append(line)
            except:
                # print(line)
                continue
    with open(fn_etype, 'w') as f:
        for line in all_lines:
            f.write(line)
    
```

```python
# remove empty files
import os    

path = "/afs/crc.nd.edu/group/dmsquare/vol2/myu2/ComparisonSentences/data/statement_scoring/news_v1"
for root,dirs,files in os.walk(path):  
    for name in files:  
        filename = os.path.join(root,name)   
        if os.stat(filename).st_size == 0:  
            print(" Removing ",filename)  
            os.remove(filename)  
```

```python
analysis_matched(dir_matched=dir_matched)
```
